```python
def death_toll(deaths, pumps, radius = 1):
    
    death_toll = []
    for i, p in pumps.iterrows():
        death_count = 0
        
        for j, d in deaths.iterrows():
            dist = ((p.x - d.x)**2 + (p.y - d.y)**2)**0.5
            if dist <= radius:
                death_count += 1
        
        death_toll.append(death_count)
    
    return death_toll
        
```

## Probabilidad condicional e independencia

**Probabilidad condicional**
$$ P(X \mid Y) ~=~ \frac{P(X \cap Y)}{P(Y)} $$

**Variables independientes**
$$ X \bot Y \implies P(X \cap Y) = P(X)P(Y) $$


$$ P(X \mid Y) ~=~ P(X) $$

## Independencia condicional

$$ X \bot Y \mid Z \implies $$

$$ P(X \cap Y \mid Z) = P(X \mid Z)P(Y \mid Z) $$

## Bayes ingenuo
Asume independencia condicional entre covariables

$$ P(C \mid X_1, \dots X_n) = \frac{P(X_1, \dots, X_n \mid C)P(C)}{P(X_1, \dots, X_n)} $$
